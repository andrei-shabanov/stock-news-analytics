```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'model': ['Dummy Classifier', 'Naive Bayes (nur Text)', 'Log Regression', 'Random Forest'],
    'weighted f1': [0.50, 0.53, 0.83, 0.82]
})

df = df.sort_values(by='weighted f1')

sns.barplot(data=df, x='model', y='weighted f1')
ax = sns.barplot(data=df, x='model', y='weighted f1')
ax.bar_label(ax.containers[0], fontsize=10);
plt.title('Weighted F1 bei den Modellen')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df2 = pd.DataFrame({
    'model': ['Dummy (MFS)', 'Dummy (Random)', 'Log Regression', 'Random Forest'],
    'accuracy': [0.50, 0.55, 0.83, 0.82],
    'weighted f1': [0.39, 0.50, 0.83, 0.82]
})

df_melted = df2.melt(id_vars='model', value_vars=['accuracy', 'weighted f1'], var_name='metric', value_name='value')

# Сортируем данные по значению метрики
df_melted = df_melted.sort_values(by='value')

# Строим график
sns.barplot(data=df_melted, x='model', y='value', hue='metric')
ax = sns.barplot(data=df_melted, x='model', y='value', hue='metric')
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=10)

plt.tight_layout()
plt.show()
```
