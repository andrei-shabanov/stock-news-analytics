```python
nasdaq_hist_data = pd.read_csv("./data/HistoricalData_1750671813123.csv")

nasdaq_hist_data.describe()

nasdaq_hist_data['change'] = np.round(
            (nasdaq_hist_data['Open'] - nasdaq_hist_data['Close/Last']) / nasdaq_hist_data['Open'], 5
        )

# based on Market-Wide Circuit Breakers of the New Yorck Stock Exchange
# https://www.nyse.com/markets/nyse/trading-info
# 1 - Strong gain (>= +20%)
# 2 - Significant gain (+13% to <20%)
# 3 - Moderate gain (+7% to <13%)
# 4 - Stable movement (-7% to <+7%)
# 5 - Moderate loss (-13% to <-7%)
# 6 - Significant loss (-20% to <-13%)
# 7 - Strong loss (< -20%)
conditions = [
    (nasdaq_hist_data['change'] >= 0.20),
    ((nasdaq_hist_data['change'] >= 0.13) & (nasdaq_hist_data['change'] < 0.20)),
    ((nasdaq_hist_data['change'] >= 0.07) & (nasdaq_hist_data['change'] < 0.13)),
    ((nasdaq_hist_data['change'] >= -0.07) & (nasdaq_hist_data['change'] < 0.07)),
    ((nasdaq_hist_data['change'] >= -0.13) & (nasdaq_hist_data['change'] < -0.07)),
    ((nasdaq_hist_data['change'] >= -0.20) & (nasdaq_hist_data['change'] < -0.13)),
    (nasdaq_hist_data['change'] < -0.20)
]

choices = [1, 2, 3, 4, 5, 6, 7]

nasdaq_hist_data['change_status'] = np.select(conditions, choices, default=np.nan)
nasdaq_hist_data['change_status'] = nasdaq_hist_data['change_status'].astype(float)
nasdaq_hist_data[nasdaq_hist_data['change'] > 0.05]
```
