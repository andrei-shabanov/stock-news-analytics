```python
nasdaq_hist_data = pd.read_csv("./data/HistoricalData_1750671813123.csv")

#nasdaq_hist_data.describe(include=[np.number])

nasdaq_hist_data['change'] = np.round(
            (nasdaq_hist_data['Open'] - nasdaq_hist_data['Close/Last']) / nasdaq_hist_data['Open'], 5
        ).map('{:.2%}'.format)

nasdaq_hist_data.sort_values(by=['change'])
print(nasdaq_hist_data)
nasdaq_hist_data.describe()
```
